```python
# Extracción de datos de ingresos de Tesla utilizando Webscraping-
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "html5lib")
```

```python
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Find the table containing the revenue data using its class
table = soup.find('table', class_='historical-data_table')

# Check if the table was found before proceeding
if table:
    for row in table.find_all("tr"):
        # Add code to process each row here
        # Example: extracting data from columns
        cols = row.find_all("td")
        if cols:
            date = cols[0].text.strip()
            revenue = cols[1].text.strip()
            tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})], ignore_index=True)
else:
    print("Table not found on the page. Check the HTML structure or your selector.")

# Display the extracted data
print(tesla_revenue.tail())
```
