```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
```

```python
print(tesla_data.head())
```

```python
# Extracción de datos de ingresos de Tesla utilizando Webscraping

html_data=requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm").text

beautiful_soup=BeautifulSoup(html_data,"html.parser")
```

```python
Tables = beautiful_soup.find_all('table')
for table in Tables:
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue_table = table
tesla_revenue=pd.DataFrame(columns=["Date","Revenue"])
tesla_revenue.head()
for row in tesla_revenue_table.find_all('tr')[1:]:
    Date=row.find_all('td')[0].text
    Revenue=row.find_all('td')[1].text
    nRow=pd.DataFrame({'Date': [Date], 'Revenue': [Revenue]})
    tesla_revenue = pd.concat([tesla_revenue, nRow], ignore_index=True)
tesla_revenue.head()
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue.tail()
```

```python
# Extracción de datos de ingresos de GameStop utilizando Webscraping

html_data_2=requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html").text
beautiful_soup_2=BeautifulSoup(html_data_2,'html.parser')
```
